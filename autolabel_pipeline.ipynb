```python
import os
import csv
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.imagenet_utils import preprocess_input
from pathlib import Path
```

```python
# Recreate the exact same model, including its weights and the optimizer
#new_model = tf.keras.models.load_model('/home/fmanzoni-lx/Documents/Compal/ResNet_Autolabel.h5')

# Show the model architecture
#new_model.summary()

# === CONFIGURAÇÕES ===
model_path = '/home/fmanzoni-lx/Documents/Compal/best_modelA.h5'
image_folder = '/home/fmanzoni-lx/Documents/Compal/Test'
output_csv = 'inferencia_resultados.csv'
target_size = (224, 224)  # Tamanho esperado pelo modelo
class_labels = ['Replay', 'Paper_Cut', 'Print']  # Suas classes

# === CARREGA O MODELO ===
model = load_model(model_path, compile=False)

# === PREPARA LISTA DE IMAGENS ===
image_paths = list(Path(image_folder).rglob('*.jpg'))  # ou *.png, etc.
```

```python
# === INFERÊNCIA E SALVA RESULTADOS ===
with open(output_csv, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['image_path', 'prediction', 'predicted_class'])

    for img_path in image_paths:
        # Carrega e pré-processa a imagem
        img = image.load_img(img_path, target_size=target_size)
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)  # Use apenas se o modelo precisa disso

        # Faz predição
        preds = model.predict(x)[0]  # vetor de probabilidades
        pred_class = class_labels[np.argmax(preds)]

        # Escreve no CSV
        writer.writerow([str(img_path), preds.tolist(), pred_class])
```
